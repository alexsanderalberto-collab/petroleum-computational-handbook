```python
#Código para encontrar as porosidades de duas amostras por meio do volume poroso e do volume total (vb)
# Dados de entrada: vp = volume poroso; vb = volume total/bulk volume
vp_a_cm3 = 12.0
vb_a_cm3 = 60.0
vp_b_cm3 = 30.0
vb_b_cm3 = 100

# Cálculo:
porosidade_a = vp_a_cm3 / vb_a_cm3
porosidade_b = vp_b_cm3 / vb_b_cm3

# Saída:
print(f"A amostra A possui porosidade de {porosidade_a} ou {porosidade_a*100}% e a amostra B de {porosidade_b} ou {porosidade_b*100}%.")
```

```python
#Código básico para descobrir a saturação de óleo a partir da saturação de água e gás
# Dados de entrada
sat_agua = 0.22
sat_gas = 0.15

# Cálculo da saturação de óleo (sat_oleo) com base na regra de ouro dos reservatórios
#Matemática: Sw + So + Sg = 1 (ou 100%)
sat_oleo = 1.0 - sat_agua - sat_gas

# Saída (imprima o valor em porcentagem utilizando uma f-string com duas casas decimais)
print(f"A saturação de óleo presente no volume poroso Vp é de: {sat_oleo:.2f} ou {sat_oleo*100}%")
```

```python
#Código para determinar o volume de óleo presente a partir do volume poroso e da saturação de óleo.
# Dados de entrada:
volume_poros_m3 = 50000.0
sat_oleo = 0.70

# Cálculo do volume de óleo
volume_oleo_m3 = volume_poros_m3 * sat_oleo

# Saída de dados
print(f"O volume de óleo presente no volume poroso é de: {volume_oleo_m3:.0f} m³.")
```

```python
# Código para relatório volumétrico de campo de petróleo
# Dados de Entrada:
nome_campo = "Campo de Libra"
area_reservatorio_m2 = 25000000.0
espessura_rocha_net_pay_m = 120.0
porosidade_media_arenito = 0.22
saturacao_agua_inicial_swi = 0.28
fator_bo_m3_m3 = 1.35

# Processamento:
volume_bulk = area_reservatorio_m2 * espessura_rocha_net_pay_m # Vb = A x h (m³)
volume_poroso_vp = volume_bulk * porosidade_media_arenito # porosidade = vp/vg
saturacao_oleo = 1.0 - saturacao_agua_inicial_swi # considerando apenas água e óleo, sem presença de gás, então So = 1 - Swi
volume_oleo = volume_poroso_vp * saturacao_oleo # volume que o óleo ocupa no volume poroso
volume_oleo_superficie_ooip = volume_oleo / fator_bo_m3_m3  # formula OOIP ou N

# Saída:
print("=" * 70)
print(f"RELATÓRIO VOLUMÉTRICO: {nome_campo}")
print("=" * 70)
print(f"Volume total da rocha (Vb): {volume_bulk:.0f} m³")
print(f"Volume de poros (Vp): {volume_poroso_vp:.0f} m³")
print(f"Saturação de óleo (So): {saturacao_oleo * 100}%")
print(f"Volume de óleo no reservatório: {volume_oleo:.0f} m³")
print(f"Volume de óleo na superfície (OOIP): {volume_oleo_superficie_ooip:.0f} ou {volume_oleo_superficie_ooip / 10**6:.0f} milhões de m³")
print("=" * 70)
```
